```python
im = getImg(df['photo_url'][4])
plt.imshow(im)
```

```python
# Erosion

kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(im,kernel,iterations = 4)
plt.imshow(erosion)
```

```python
ims = [getImg(df['photo_url'][i])for i in range(5)]
```

```python
# Edges
laps = [cv2.Laplacian(im,cv2.CV_64F,ksize=7) for im  in ims]
plt.imshow(laps[4])
print [np.mean(im) for im in laps] 
print [np.mean(im) for im in  ims]

laplacian = cv2.Sobel(im,cv2.CV_64F,1,0,ksize=13)
plt.imshow(laplacian)
```

```python
# Edges

laps = [cv2.Canny(im,100,200) for im  in ims]
plt.imshow(laps[4])
print [np.mean(im) for im in laps] 
print [np.mean(im) for im in  ims]

edges = cv2.Canny(im,100,100)
plt.imshow(edges)
```
